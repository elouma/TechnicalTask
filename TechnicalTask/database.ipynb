```python
import sqlite3
import csv

conn = sqlite3.connect("Broker.db")
cur = conn.cursor()

cur.execute(""" 
CREATE TABLE Purchase_Order
('Order date' date,
'Purchase Order' varchar(36) NOT NULL,
  'Customer ID' varchar(32), 
  'Customer Location' varchar(32),
'Order Req Amt (Ton)' int NOT NULL,
FOREIGN KEY('Customer ID') REFERENCES Customer('Customer ID'),
 PRIMARY KEY('Purchase Order')
); """)

cur.execute(""" 
CREATE TABLE Supplier
('Fullfilled By ID' varchar(36) NOT NULL,
Fullfilled By Location varchar(30),
 PRIMARY KEY('Fullfilled By ID')
); """)


cur.execute(""" 
CREATE TABLE Order_Product
('Purchase Order' varchar(36) NOT NULL,
'Fullfilled By ID' varchar(36) NOT NULL,
'Fullfilled By Location' varchar(32),
'Supplied Amt (Ton)' int NOT NULL,
'Cost Of Delivery ($)' int NOT NULL,
FOREIGN KEY('Purchase Order') REFERENCES Purchase_Order('Purchase Order'),
FOREIGN KEY('Fullfilled By ID') REFERENCES Supplier('Fullfilled By ID'),
FOREIGN KEY('Fullfilled By Location') REFERENCES Supplier('Fullfilled By Location'),
PRIMARY KEY ('Purchase Order','Fullfilled By ID')
); """)

conn.commit()

purchase = open('Purchase_Order.csv')
order = open ('Order_Product.csv')
supplier = open('Supplier.csv')


with open('/Users/Yahnik/Desktop/Blueprism/Purchase_Order.csv', 'r') as file:
    no_records = 0
    for row in file:
        cur.execute(
           "INSERT INTO Purchase_Order VALUES(?,?,?,?,?)", row.split(","))
        conn.commit()
        no_records += 1
print('\n{} Records Transferred'.format(no_records))


with open('/Users/Yahnik/Desktop/Blueprism/Order_Product.csv', 'r') as file:
    no_records = 0
    for row in file:
        cur.execute(
           "INSERT INTO Order_Product VALUES(?,?,?,?,?)", row.split(","))
        conn.commit()
        no_records += 1
print('\n{} Records Transferred'.format(no_records))


with open('/Users/Yahnik/Desktop/Blueprism/Supplier.csv', 'r') as file:
    no_records = 0
    for row in file:
        cur.execute(
           "INSERT INTO Supplier VALUES(?,?)", row.split(","))
        conn.commit()
        no_records += 1

print('\n{} Records Transferred'.format(no_records))
conn.close()
```
